```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
import spacy
import faiss
from sklearn.metrics.pairwise import cosine_similarity
import re
from typing import Dict, List, Tuple
import json

print("✓ All imports successful!")
```

```python
print("Loading Sentence-BERT model (all-MiniLM-L6-v2)...")
model = SentenceTransformer('all-MiniLM-L6-v2')

print("Loading spaCy NLP model...")
nlp = spacy.load('en_core_web_sm')

print("✓ Models loaded successfully!")
```

```python
class ResumeParser:
    """Parse and extract information from resumes"""

    def __init__(self):
        self.nlp = nlp

    def extract_skills(self, text):
        """Extract skills from resume text"""
        skill_keywords = {
            'programming': ['python', 'java', 'javascript', 'c++', 'c#', 'go', 'rust', 'php', 'swift'],
            'web': ['react', 'angular', 'vue.js', 'html', 'css', 'node.js', 'django', 'flask'],
            'databases': ['sql', 'mysql', 'postgresql', 'mongodb', 'dynamodb', 'redis', 'cassandra'],
            'cloud': ['aws', 'azure', 'gcp', 'kubernetes', 'docker'],
            'ml': ['tensorflow', 'pytorch', 'scikit-learn', 'keras', 'machine learning', 'deep learning'],
            'tools': ['git', 'jenkins', 'jira', 'linux', 'unix', 'docker', 'kubernetes']
        }

        found_skills = {}
        text_lower = text.lower()

        for category, skills in skill_keywords.items():
            found_skills[category] = [skill for skill in skills if skill in text_lower]

        return found_skills

    def extract_experience_years(self, text):
        """Extract years of experience"""
        experience_patterns = [
            r'(\d+)\+?\s*years?\s+(?:of\s+)?experience',
            r'(?:experience|worked|employed).*?(\d+)\+?\s*years?'
        ]

        for pattern in experience_patterns:
            match = re.search(pattern, text, re.IGNORECASE)
            if match:
                return int(match.group(1))
        return 0

    def extract_education(self, text):
        """Extract education information"""
        edu_patterns = {
            'PhD': r'ph\.?d|doctorate',
            'Master': r"master'?s|m\.s|m\.a|m\.tech",
            'Bachelor': r"bachelor'?s|b\.s|b\.a|b\.tech|b\.e",
            'Diploma': r'diploma|certification'
        }

        found_education = []
        text_lower = text.lower()

        for degree, pattern in edu_patterns.items():
            if re.search(pattern, text_lower):
                found_education.append(degree)

        return found_education

    def parse_resume_text(self, text):
        """Parse resume and extract all information"""
        parsed_data = {
            'full_text': text,
            'skills': self.extract_skills(text),
            'experience_years': self.extract_experience_years(text),
            'education': self.extract_education(text),
            'word_count': len(text.split()),
            'summary': text[:300] + "..." if len(text) > 300 else text
        }
        return parsed_data

print("✓ ResumeParser class created!")
```

```python
class SemanticMatcher:
    """Match resumes to job descriptions using semantic similarity"""

    def __init__(self, model):
        self.model = model
        self.resume_embeddings = None
        self.job_embeddings = None
        self.resumes = []
        self.jobs = []
        self.index = None
        self.resume_metadata = []
        self.job_metadata = []

    def add_resumes(self, resume_texts: List[str], metadata: List[Dict] = None):
        """Add resumes and create embeddings"""
        self.resumes = resume_texts
        self.resume_metadata = metadata if metadata else [{'id': i} for i in range(len(resume_texts))]

        print(f"Creating embeddings for {len(resume_texts)} resumes...")
        self.resume_embeddings = self.model.encode(resume_texts, show_progress_bar=True)

        # Create FAISS index
        dimension = self.resume_embeddings.shape[1]
        self.index = faiss.IndexFlatL2(dimension)
        self.index.add(self.resume_embeddings.astype(np.float32))

        print(f"✓ FAISS index created with {len(resume_texts)} resumes\n")

    def add_jobs(self, job_texts: List[str], metadata: List[Dict] = None):
        """Add job descriptions"""
        self.jobs = job_texts
        self.job_metadata = metadata if metadata else [{'id': i} for i in range(len(job_texts))]

        print(f"Encoding {len(job_texts)} job descriptions...")
        self.job_embeddings = self.model.encode(job_texts, show_progress_bar=True)
        print("✓ Job embeddings created\n")

    def find_top_matches(self, job_index: int, top_k: int = 5) -> List[Tuple[int, float]]:
        """Find top matching resumes for a job"""
        if self.index is None or job_index >= len(self.jobs):
            raise ValueError("Invalid job index or no resumes added")

        job_embedding = self.job_embeddings[job_index:job_index+1].astype(np.float32)
        distances, indices = self.index.search(job_embedding, min(top_k, len(self.resumes)))

        # Convert L2 distances to similarity scores (0-1)
        scores = 1 / (1 + distances[0])

        # Corrected line: use indices[0] to get the 1D array of actual indices
        return [(int(idx), float(score)) for idx, score in zip(indices[0], scores)]

    def get_similarity_matrix(self) -> np.ndarray:
        """Get similarity matrix between all resumes and jobs"""
        similarity = cosine_similarity(self.job_embeddings, self.resume_embeddings)
        return similarity

    def calculate_match_score(self, resume_text: str, job_text: str) -> float:
        """Calculate single match score"""
        resume_emb = self.model.encode(resume_text).reshape(1, -1)
        job_emb = self.model.encode(job_text).reshape(1, -1)
        similarity = cosine_similarity(resume_emb, job_emb)[0][0]
        return float(similarity)

print("✓ SemanticMatcher class created!")
```

```python
print("Initializing TalentSync Matcher...\n")

matcher = SemanticMatcher(model)

print("Adding resumes to matcher...")
matcher.add_resumes(sample_resumes, resume_metadata)

print("Adding job descriptions to matcher...")
matcher.add_jobs(sample_jobs, job_metadata)

print("=" * 70)
print("✓ SYSTEM READY FOR MATCHING")
print("=" * 70)
```

```python
print("\nGenerating Match Results...\n")
print("=" * 70)
print("RESUME-TO-JOB MATCHING RESULTS")
print("=" * 70)

results = {}
for job_idx, job_text in enumerate(sample_jobs):
    job_title = job_metadata[job_idx]['title']
    job_company = job_metadata[job_idx]['company']

    matches = matcher.find_top_matches(job_idx, top_k=len(sample_resumes))

    print(f"\n📍 {job_title}")
    print(f"   Company: {job_company}")
    print(f"   {'─' * 65}")

    for rank, (resume_idx, score) in enumerate(matches, 1):
        candidate_name = resume_metadata[resume_idx]['name']
        candidate_pos = resume_metadata[resume_idx]['position']
        match_percentage = score * 100

        # Visual indicator
        bar = "█" * int(match_percentage / 5) + "░" * (20 - int(match_percentage / 5))

        print(f"   {rank}. {candidate_name:20} | {candidate_pos:25} | [{bar}] {match_percentage:.1f}%")

    results[f"Job_{job_idx}"] = matches

print("\n" + "=" * 70)
```

```python
print("\n\n" + "=" * 70)
print("DETAILED RESULTS TABLE")
print("=" * 70)

detailed_results = []
for job_idx in range(len(sample_jobs)):
    matches = matcher.find_top_matches(job_idx, top_k=len(sample_resumes))
    for rank, (resume_idx, score) in enumerate(matches, 1):
        detailed_results.append({
            'Job_ID': job_idx + 1,
            'Job_Title': job_metadata[job_idx]['title'],
            'Company': job_metadata[job_idx]['company'],
            'Resume_ID': resume_idx + 1,
            'Candidate_Name': resume_metadata[resume_idx]['name'],
            'Position': resume_metadata[resume_idx]['position'],
            'Match_Score_%': f"{score * 100:.1f}%",
            'Rank': rank
        })

results_df = pd.DataFrame(detailed_results)
print(results_df.to_string(index=False))
```

```python
print("\n\n" + "=" * 70)
print("MATCHING STATISTICS")
print("=" * 70)

match_scores = []
for job_idx in range(len(sample_jobs)):
    matches = matcher.find_top_matches(job_idx, top_k=len(sample_resumes))
    match_scores.extend([score for _, score in matches])

match_scores = np.array(match_scores)

print(f"\nTotal Matches Evaluated: {len(match_scores)}")
print(f"Average Match Score: {match_scores.mean() * 100:.1f}%")
print(f"Highest Match Score: {match_scores.max() * 100:.1f}%")
print(f"Lowest Match Score: {match_scores.min() * 100:.1f}%")
print(f"Std Deviation: {match_scores.std() * 100:.1f}%")
print(f"\nMatches above 70%: {len(match_scores[match_scores > 0.7])}")
print(f"Matches above 50%: {len(match_scores[match_scores > 0.5])}")
print(f"Matches below 50%: {len(match_scores[match_scores <= 0.5])}")
```

```python
print("\n\n" + "=" * 70)
print("SAVING RESULTS")
print("=" * 70)

# Calculate similarity matrix
similarity_matrix = matcher.get_similarity_matrix()
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=[job['title'] for job in job_metadata],
    columns=[resume['name'] for resume in resume_metadata]
)

# Create best matches summary DataFrame
best_matches = results_df.loc[results_df.groupby('Job_ID')['Rank'].idxmin()]
summary_df = best_matches.drop(columns=['Job_ID', 'Resume_ID', 'Rank']).reset_index(drop=True)

# Save similarity matrix
similarity_df.to_csv('similarity_matrix.csv')
print("✓ Similarity matrix saved as 'similarity_matrix.csv'")

# Save detailed results
results_df.to_csv('matching_results.csv', index=False)
print("✓ Detailed results saved as 'matching_results.csv'")

# Save summary
summary_df.to_csv('best_matches_summary.csv', index=False)
print("✓ Best matches saved as 'best_matches_summary.csv'")

print("\n✓ All processing complete! Ready for Streamlit deployment.")
print("\nFiles available for download in Colab environment.")
```
